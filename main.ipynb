```python
from datasets import load_from_disk
import matplotlib.pyplot as plt
import pandas as pd

dataset = load_from_disk("./dataset_dw")

df = dataset.to_pandas()

post_df = pd.json_normalize(df["post"])
```

```python

top_20 = post_df['submolt.display_name'].value_counts().head(20)

plt.figure(figsize=(12, 6))
top_20.plot(kind='bar', color='steelblue')
plt.title('Top 20 Submolt per numero di post', fontsize=14, fontweight='bold')
plt.xlabel('Submolt', fontsize=12)
plt.ylabel('Numero di post', fontsize=12)
plt.xticks(rotation=-45, ha='left')
plt.show()
```

```python
toxic_level = df["toxic_level"].value_counts()

plt.figure(figsize=(12, 6))
toxic_level.plot(kind='bar')
plt.title('Distribuzione toxic_level', fontsize=14, fontweight='bold')
plt.xlabel('Toxic Level', fontsize=12)
plt.ylabel('Numero di post', fontsize=12)
plt.xticks(rotation=0)
plt.show()
```

```python

df["text_len"] = post_df["content"].str.len()
print(df["text_len"].describe())
```

```python
plt.figure(figsize=(12, 6))
df["text_len"].hist(bins=300)
plt.title('Distribuzione lunghezza post', fontsize=14, fontweight='bold')
plt.xlabel('Lunghezza del testo', fontsize=12)
plt.ylabel('Numero di post', fontsize=12)
plt.xlim(0, 8000) 
plt.show()
```

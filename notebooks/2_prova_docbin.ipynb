```python
import spacy
from spacy.tokens import DocBin
import pandas as pd
from collections import Counter

nlp = spacy.load('en_core_web_sm')
doc_bin = DocBin().from_disk('../data/processed/docs_prova.docbin')
docs = list(doc_bin.get_docs(nlp.vocab))
```

## Lunghezza media dei post in token

```python

token_count = [len(doc) for doc in docs]
stats = pd.Series(token_count, name="Lenght in tokens").describe()
print(stats)
```

## Media TTR nel dataset

```python
    
ttr_list = []
for doc in docs:
    unique_tokens = len(set([token.text.lower() for token in doc]))
    ttr = unique_tokens / len(doc) if len(doc) > 0 else 0
    ttr_list.append(ttr)
media_ttr = sum(ttr_list)/len(ttr_list)
print("Media TTR:", media_ttr)
```

```python
pos_global = Counter()
for doc in docs:
    for token in doc:
        pos_global[token.pos_] += 1

print(pos_global.most_common())
```
